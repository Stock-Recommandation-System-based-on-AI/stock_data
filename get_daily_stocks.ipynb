```python
def get_daily_stock_data(itmes):
    import datetime    
    instStockChart = win32com.client.Dispatch("CpSysDib.StockChart")
    nCpCybos = win32com.client.Dispatch("CpUtil.CpCybos")
    
    column_dailychart = ['code', 'name', 'date', 'open', 'high', 'low', 'close',]
    
    instStockChart.SetInputValue(1, ord('1'))
    instStockChart.SetInputValue(5, (0, 2, 3, 4, 5))
    instStockChart.SetInputValue(6, ord('D'))
    instStockChart.SetInputValue(9, ord('1'))
    
    rows = list()
    for code, name in itmes.items():
        row = list(range(len(column_dailychart)))
        # 현재가 객체 구하기
        instStockChart.SetInputValue(0, code)
        
        t = datetime.date.today()
        today = t.strftime("%Y%m%d")
        instStockChart.SetInputValue(3, today)
        instStockChart.BlockRequest()

        numData = instStockChart.GetHeaderValue(3)
        numField = instStockChart.GetHeaderValue(1)
        # GetDataValue
        for i in range(numData):
            row[0] = code
            row[1] = name
            row[2] = instStockChart.GetDataValue(0, i)# 날짜
            row[3] = instStockChart.GetDataValue(1, i) # 시가
            row[4] = instStockChart.GetDataValue(2, i) # 고가
            row[5] = instStockChart.GetDataValue(3, i) # 저가
            row[6] = instStockChart.GetDataValue(4, i) # 종가
            rows.append(list(row))
    stockitems_dailychart = pd.DataFrame(rows, columns=column_dailychart)
    return stockitems_dailychart
```

```python
market_daily_chart = get_daily_stock_data(market_itmes)
market_daily_chart
```

```python
stocks_daily_chart = get_daily_stock_data(stock_items)
stocks_daily_chart
```

```python
market = pd.concat([market, market_daily_chart])
market = market.sort_values(["name", "date"])
market
```

```python
stocks = pd.concat([stocks, stocks_daily_chart])
stocks = stocks.sort_values(["name", "date"])
stocks
```

```python
kospi31_stocks = pd.concat([market, stocks])
kospi31_stocks.reset_index(drop=True, inplace=True)
kospi31_stocks
```

```python
kospi31_stocks.to_csv("./data/kospi31-2003-2021.csv", index=False)
```

```python
def preprocess(stock_chart):
    z_feature = stock_chart[["code", "name", "date"]].values.tolist()
    stock_chart = stock_chart.values.tolist()
    stock_first_index = 0
    
    for idx, daily in enumerate(stock_chart):
        z_feature[idx].append(daily[3] / daily[6] - 1)    # open
        z_feature[idx].append(daily[4] / daily[6] - 1)    # high
        z_feature[idx].append(daily[5] / daily[6] - 1)    # low
        
        if stock_chart[idx-1][1] != daily[1]:
            s_idx = idx
            for i in range(7):
                z_feature[idx].append(0) # close
            continue
        else:
            z_feature[idx].append(daily[6] / stock_chart[idx-1][6] - 1) # close
        
        for k in [5, 10, 15, 20, 25, 30]: # z_dk
            if (idx-s_idx)-(k-1) < 0:
                sum_close = stock_chart[s_idx:idx+1]
            else:
                sum_close = stock_chart[idx-(k-1):idx+1]

            sum_close = [chart[6] for chart in sum_close]
            sum_close = sum(sum_close) / (len(sum_close) * daily[6]) - 1
            z_feature[idx].append(sum_close)

    return pd.DataFrame(z_feature, columns=["code", "name", "date", "open", "high", "low", "close",
                                            "z_d5", "z_d10", "z_d15", "z_d20", "z_d25", "z_d30"])
```

```python
preprocessed_kospi31_daily_stock_chart = preprocess(kospi31_stocks)
preprocessed_kospi31_daily_stock_chart
```

```python
preprocessed_kospi31_daily_stock_chart.to_csv("./data/preprocessed_kospi31_daily_stock_chart.csv", index=False)
```
